```python
import pandas as pd
from matplotlib import pyplot as plt
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s',
handlers=[
        logging.StreamHandler(),  # Stream logs to the console
        logging.FileHandler('piechart.log')  # Store logs in a file
    ])
    # Create a logger
logger = logging.getLogger('my_logger')
```

1. Write a Python programming to create a pie chart of the popularity of programming
Languages.

```python
try:
    prog_lang=['C++', 'C#', 'JavaScript', 'PHP', 'Python', 'Java']
    popular=[6.7, 7.7, 8, 8.8, 17.6, 22.2]

    plt.pie(popular,labels=prog_lang,autopct='%1.1f%%',wedgeprops={'edgecolor':'white'},startangle=90)
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

2. Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.

```python
try:
    prog_lang=['C++', 'C#', 'JavaScript', 'PHP', 'Python', 'Java']
    popular=[6.7, 7.7, 8, 8.8, 17.6, 22.2]

    plt.pie(popular,labels=prog_lang,autopct='%1.1f%%',wedgeprops={'edgecolor':'white'},startangle=90)
    plt.title("popularity of programming Languages.")
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

3. Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics.

```python
try:
    # Read data from CSV file
    df = pd.read_csv('medal.csv')

    # Extract country and gold medal columns
    countries = df['country']
    gold_medals = df['gold_medal']
    color = ['red', 'green', 'blue', 'orange', 'purple']
    # Create the pie chart
    text,country_labels,percentage=plt.pie(gold_medals, labels=countries, autopct='%1.1f%%', startangle=90,colors=color)
    for index,label in enumerate(country_labels):
        label.set_color(color[index])
    # Set the title
    plt.title('Gold Medal Achievements - 2016 Summer Olympics')

    # Show the plot
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```
