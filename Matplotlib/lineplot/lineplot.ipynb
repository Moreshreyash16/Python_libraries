14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python
try:
    # read data from csv file using pandas
    df = pd.read_csv('alphainc.csv', sep=',' ,parse_dates=True, index_col=0)

    print(df)

    plt.plot(df,linestyle='-', linewidth=.5,color='blue')
    
    plt.title("closing value of Alphabet Inc.")
    plt.xticks(rotation=45)
    plt.grid(True,which='minor')
    plt.tick_params(axis='both', which='minor', bottom=False, left=False)
    plt.show()
    logger.info("Read data from csv and plotted successfully")

except Exception as e:
    logger.error(e)
```

15. Write a Python program to create multiple plots.

```python
try:
    x = np.linspace(0,6,100)
    y1 = np.sin(x)
    y2 = np.cos(x)


    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

    ax1.plot(x, y1, color='blue', label='Sin(x)')
    ax1.set_title('Plot 1')
    ax1.set_xlabel('x')
    ax1.set_ylabel('sin(x)')
    ax1.legend()
    ax1.grid(True)

    ax2.plot(x, y2, color='red', label='Cos(x)')
    ax2.set_title('Plot 2')
    ax2.set_xlabel('x')
    ax2.set_ylabel('cos(x)')
    ax2.grid(True)
    ax2.legend()

    plt.tight_layout()

    plt.show()
except Exception as e:
    logger.error(e)
```
