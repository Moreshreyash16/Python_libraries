5. Write a Python programming to display a bar chart of the popularity of programming
Languages. Attach a text label above each bar displaying its popularity (float value).

```python
try:
    # Define the programming languages and their popularity values
    prog_lang = ['C++', 'C#', 'JavaScript', 'PHP', 'Python', 'Java']
    popular = [6.7, 7.7, 8, 8.8, 17.6, 22.2]

    # Create a bar chart with programming languages as x-axis and popularity as y-axis
    plt.bar(prog_lang, popular, label=popular)

    # Set the title, x-axis label, and y-axis label of the plot
    plt.title("Popularity of Programming Languages")
    plt.xlabel("Programming Languages")
    plt.ylabel("Popularity")

    # Add text labels above each bar showing the popularity value
    for i in range(len(prog_lang)):
        plt.text(i, popular[i], str(popular[i]), ha='center', va='bottom')

    # Display the plot
    plt.show()

    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

9. Write a Python programming to display a bar chart of the popularity of programming
Languages. Increase bottom margin.

```python

# Programming languages
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']

# Popularity values
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create a figure and axes
fig, ax = plt.subplots()

# Create a bar chart
ax.bar(languages, popularity)

# Set the title and labels
ax.set_title('Popularity of Programming Languages')
ax.set_xlabel('Programming Languages')
ax.set_ylabel('Popularity')

# Adjust the bottom margin
plt.subplots_adjust(bottom=0.1)

# Display the plot
plt.show()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.

```python

# Sample data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]


width=0.35
# Set the positions for the bars
positions_men = np.arange(len(groups))
positions_women = positions_men+width 

# Create the bar plot for men
plt.bar(positions_men, means_men,width=width, label='Men')

# Create the bar plot for women
plt.bar(positions_women, means_women,width=width, label='Women')

# Set the x-axis tick labels
plt.xticks(positions_men , groups)

# Set the title and labels
plt.title('Scores by Group and Gender')
plt.xlabel('Groups')
plt.ylabel('Scores')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure.

```python
try:
    # Sample data
    means = [0.2474, 0.1235, 0.1737, 0.1824]
    std_devs = [0.3314, 0.2278, 0.2836, 0.2645]
    labels = ['A', 'B', 'C', 'D']

    # Set the positions for the bars
    x_pos = np.arange(len(labels))

    # Create the bar plot
    plt.bar(x_pos, means, yerr=std_devs, align='center', alpha=0.8, capsize=10)

    # Set the x-axis tick labels
    plt.xticks(x_pos, labels)

    # Set the title and labels
    plt.title('Mean Velocity with Error Bars')
    plt.xlabel('Categories')
    plt.ylabel('Mean Velocity')

    # Show the plot
    plt.show()
except Exception as e:
    logger.error(e)

```

13. Write a Python program to create bar plots with errorbars on the same figure.Attach a text
label above each bar displaying men means (integer value).

```python
try:
    # Sample data
    means = [0.2474, 0.1235, 0.1737, 0.1824]
    std_devs = [0.3314, 0.2278, 0.2836, 0.2645]
    labels = ['A', 'B', 'C', 'D']

    # Set the positions for the bars
    x_pos = np.arange(len(labels))

    # Create the bar plot
    plt.bar(x_pos, means, yerr=std_devs, align='center', alpha=0.8, capsize=10)

    # Set the x-axis tick labels
    plt.xticks(x_pos, labels)

    for i,v in enumerate(means):
        plt.text(i, v, str(v), ha='left', va='bottom')    
    # Set the title and labels
    plt.title('Mean Velocity with Error Bars')
    plt.xlabel('Categories')
    plt.ylabel('Mean Velocity')

    # Show the plot
    plt.show()
except Exception as e:
    logger.error(e)

```
