```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s',
handlers=[
        logging.StreamHandler(),  # Stream logs to the console
        logging.FileHandler('seaborn.log')  # Store logs in a file
    ])
    # Create a logger
logger = logging.getLogger('my_logger')
```

<b>Seaborn Bar plot</b>
1. Write a program to draw bar plot of sex against survived for a dataset given in the url

```python
try:
    titanic=sns.load_dataset('titanic')
    sns.countplot(data=titanic,x='sex',hue='survived')
    plt.title("survived or not")
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

<b>Seaborn point plot</b>

 2. Write a program to draw a point plot for sex against survived for a dataset given in url

```python
try:
    titanic=sns.load_dataset('titanic')
    sns.pointplot(data=titanic, y='survived', x='sex')
    plt.title("survived or not")
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

<b>Seaborn scatter plot</b>

3. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url


```python
try:
    tips=sns.load_dataset('tips')
    sns.scatterplot(data=tips, x='day', y='total_bill')
    plt.title("Bill amount")
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

<b>Seaborn Violin Plot</b>

4. Write a program to draw a violin plot of sex against total_bill for a given dataset


```python
try:
    sns.violinplot(data=tips, x='sex', y='total_bill')
    plt.title("Bill amount")
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

<b>Seaborn BoxPlot</b>

6. Write a program to draw box plot of life expectancy by continent for a data set given in a
url

```python
try:
    url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
    gapminder = pd.read_csv(url)
    sns.boxplot(data=gapminder, x='continent', y='lifeExp')
    plt.title("life expectancy")
    plt.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```
