```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s',
handlers=[
        logging.StreamHandler(),  # Stream logs to the console
        logging.FileHandler('plotly.log')  # Store logs in a file
    ])
    # Create a logger
logger = logging.getLogger('my_logger')
```

1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
try:
    x = np.random.randn(1000)  # Random X coordinates
    y = np.random.randn(1000)  # Random Y coordinates

    # Create a DataFrame with x and y coordinates
    data = {'x': x, 'y': y}
    df = pd.DataFrame(data)

    # Create the scatter plot
    fig = px.scatter(df, x='x', y='y', color='x')

    # Show the plot
    fig.show()
    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
try:
    # Generate random x and y coordinates
    np.random.seed(0)
    x = np.random.randn(100)  # Random X coordinates
    y = np.random.randn(100)  # Random Y coordinates

    # Line plot
    fig = px.line(x=x, y=y)
    fig.update_layout(title='Line Plot', xaxis_title='X', yaxis_title='Y')
    fig.show()

    # Scatter plot
    fig = px.scatter(x=x, y=y)
    fig.update_layout(title='Scatter Plot', xaxis_title='X', yaxis_title='Y')
    fig.show()

    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
try:
    
    x = np.random.randn(500)  # Random X coordinates
    y = np.random.randn(500)  # Random Y coordinates

    # Take the absolute value of y coordinates
    y_abs = np.abs(y)

    # Create the scatter plot with style
    fig = px.scatter(x=x, y=y, color=x, size=y_abs, opacity=0.8)

    # Update the layout
    fig.update_layout(
        title='Scatter Plot',
        xaxis_title='X',
        yaxis_title='Y',
        showlegend=True
    )

    # Show the plot
    fig.show()

    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover

```python
ds=pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv")
ds.head(3)
```

```python
# Read the dataset from the provided URL
url = 'https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv'
df = pd.read_csv(url)

try:
    # Create the scatter plot with data labels on hover
    fig = px.scatter(df, x='Population', y='State',hover_data=['Rank','Postal'],color='Population',size='Population')
    fig.update_layout(
        title='USA States'
    )
    fig.show()

    logger.info("Program executed successfully!")
except Exception as e:
    logger.error(e)
```
